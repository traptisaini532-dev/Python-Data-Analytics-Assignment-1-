 # 1. Explain: Data Cleaning & Preprocessing 
• Load the dataset and check for missing values. 
• Fill missing Age values with median or mean. 
• Handle missing Embarked values by filling with the most common port. • Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes.

```python
# • Load the dataset and check for missing values. 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\Data Anayltic\Numpy\titanic_dataset.csv")

# Show first 5 rows
print(df.head())

# Check missing values
print(df.isnull().sum())
```

```python
# Fill missing Age values with median or mean.

w=df['Age'].fillna(df['Age'].median())
print(w)
```

```python
# Handle missing Embarked values by filling with the most common port.

q=df['Embarked'].fillna(df['Embarked'].mode()[0])
print(q)
```

```python
# Drop unnecessary columns like Name, Ticket, Cabin for visualization purposes.

a=df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
print(a)
```

 # 2. Univariate Analysis 
• Plot the distribution of Age using a histogram and a KDE plot. 
• Plot the count of passengers by Sex using a bar chart. 
• Show the distribution of Fare using a boxplot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the distribution of Age using a histogram and a KDE plot. 
plt.figure()
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#  Plot the count of passengers by Sex using a bar chart.
plt.figure()
sns.countplot(x="Sex", data=df)
plt.title("Count of Passengers by Sex")
plt.show()
```

```python
#  Show the distribution of Fare using a boxplot.
plt.figure()
sns.boxplot(x=df["Fare"])
plt.title("Fare Distribution")
plt.show()
```

  # 3. Bivariate Analysis (Two Variable Analysis)
   • Compare survival rate across Sex using a countplot. 
   • Compare Pclass vs Fare using a boxplot. 
   • Visualize Age vs Fare using a scatter plot, color-coded by Survived.

```python
# • Compare survival rate across Sex using a countplot. 
plt.figure()
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
# • Compare Pclass vs Fare using a boxplot.
plt.figure()
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Pclass vs Fare")
plt.show()
```

```python
#  • Visualize Age vs Fare using a scatter plot, color-coded by Survived.
plt.figure()
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```
