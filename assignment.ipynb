# Q2. Write a program to find the frequency of elements in a list

```python
lst=[2,4,5,3,4,3,2,2,4]
dict={}
for i in lst:
    dict[i]=dict.get(i,0)+1
print(dict)
```


# Q4. Write a program to merge two dictionaries?

```python
dict1={"a":1,"b":2,"c":3}
dict2={"q":5,"w":6,"r":8}
result={}
for k in dict1.keys()|dict2.keys():
    result[k]=max(dict1.get(k,0),dict2.get(k,0))
print(result)  
```

# Q5. Using a dictionary comprehension, create a dictionary from a list of integers where keys are numbers and values are their cubes, but only for odd numbers?

```python
num1=[3,5,8,9,7,1,0]
d={n:n*n*n for n in num1 if n% 2!=0}
print(d)
```
